# Ejemplo básico: Verificación de la involución canónica

Este notebook demuestra el uso del pipeline incluido en `canonical-involution-benchmarks`.
Incluye:
- Definición simbólica del sistema
- Prolongación de primer orden
- Aplicación de la involución canónica
- Cálculo de discrepancia simbólica

```python
from prolongation import prolong
from involution import apply_canonical_involution
from discrepancy import compute_discrepancy

from sympy import symbols, Function

x, y = symbols('x y')
u = Function('u')(x, y)

example_system = {
    "u_x": u.diff(x),
    "u_y": u.diff(y)
}

example_system
```

```python
# Prolongación simbólica
prolonged = prolong(example_system)
prolonged
```

```python
# Aplicación de la involución canónica
involuted = apply_canonical_involution(prolonged)
involuted
```

```python
# Cálculo de discrepancia simbólica
discrepancy = compute_discrepancy(prolonged, involuted)
discrepancy
```
